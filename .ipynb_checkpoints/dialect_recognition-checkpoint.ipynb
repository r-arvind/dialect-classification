```python
from python_speech_features import mfcc
from python_speech_features import delta
import matplotlib.pyplot as plt
from scipy.io import wavfile as wav
from scipy.fftpack import fft
import numpy as np
import pandas as pd
```

```python
column1 = ['mf1','mf2','mf3','mf4','mf5','mf6','mf7','mf8','mf9','mf10']
column3 = ['mx1','mx2','mx3','mx4','mx5','mx6','mx7','mx8','mx9','mx10']
column2 = ['delta_1','delta_2','delta_3','delta_4','delta_5','delta_6','delta_7','delta_8','delta_9','delta_10']
column4 = ['mn1','mn2','mn3','mn4','mn5','mn6','mn7','mn8','mn9','mn10']
column5 = ['delta_1_mx','delta_2_mx','delta_3_mx','delta_4_mx','delta_5_mx','delta_6_mx','delta_7_mx','delta_8_mx','delta_9_mx','delta_10_mx']
column6 =  ['delta_1_mn','delta_2_mn','delta_3_mn','delta_4_mn','delta_5_mn','delta_6_mn','delta_7_mn','delta_8_mn','delta_9_mn','delta_10_mn']
#column7 =  ['delta_1_m','delta_2_mx','delta_3_mx','delta_4_mx','delta_5_mx','delta_6_mx','delta_7_mx','delta_8_mx','delta_9_mx','delta_10_mx']
```

```python
#Data Reading for 1st dialect

total_data_1 = [] # we will append data of ist dialect here from every file to i

for i in range(1,68):
    [fs,data] = wav.read("Read_Up/IDR1/{}.wav".format(i))    
    #extracting mfcc features 
    mfcc_data = mfcc(data,fs,numcep = 10)
    
    mfcc_data_max = np.max(mfcc_data,axis = 0)[np.newaxis]
    mfcc_data_min = np.min(mfcc_data,axis = 0)[np.newaxis]
    mfcc_data_mean = np.mean(mfcc_data,axis = 0)[np.newaxis]
    
    mfcc_data_mean = pd.DataFrame(data = mfcc_data_mean,columns = column1)
    mfcc_data_max = pd.DataFrame(data = mfcc_data_max,columns = column3)
    mfcc_data_min = pd.DataFrame(data = mfcc_data_min,columns = column4)
    #extracting delta_feat 
    
    delta_feat = delta(mfcc_data,4)
    delta_feat_mean = np.mean(delta_feat,axis = 0)[np.newaxis]
    delta_feat_max =  np.max(mfcc_data,axis = 0)[np.newaxis]
    delta_feat_min =  np.min(mfcc_data,axis = 0)[np.newaxis]
    
    delta_feat_mean = pd.DataFrame(data = delta_feat_mean,columns = column2)
    delta_feat_max = pd.DataFrame(data = delta_feat_max,columns = column5)
    delta_feat_min = pd.DataFrame(data = delta_feat_min,columns = column6)
    
    #concating all the features 
    dialect_1= pd.concat([mfcc_data_mean,mfcc_data_min,mfcc_data_max,delta_feat_mean,delta_feat_max,delta_feat_min],axis = 1)
    dialect_1['outcome'] = [1 for i in range(len(dialect_1))]
    total_data_1.append(dialect_1)
#including energy as also as one feature
```
